### Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

A DataFrame is a 2-dimensional data structure that can store data of different types (including characters, integers, floating point values, categorical data and more) in columns. It is similar to a spreadsheet, a SQL table.

DataFrame is two dimentional, whereas Series is one dimentional. Each column in a DataFrame is a Series. Dataframe has column label, but a series does not have a column label.

<b><u>Example:</u></b>

If you consider "series1" series and "df1" dataframe, you can see "series1" does not have any column level and it is one dimentional and "df1" dataframe does have column levels for each columns.

<pre>
To access "23" in "series1" : series1[4]
To access "27" in "df1" : df1["Age"][2]
</pre>

```python
print(series1)
print(df1)
```

```python
print(series1[4])
print(df1["Age"][2])
```

### Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

read_csv() and to_csv() are to commonly used methods to read data from a .csv file, even from a URL link which has CSV data into a DataFrame and write data from a dataframe into a .csv file. Below example given of these two functions.



```python
td = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv")
td
```

```python
# after executing the below code, you will be able to see on the same directory, "titanic.csv" file 
# has been created with the data in the DataFrame.

td.to_csv("titanic.csv")
```
