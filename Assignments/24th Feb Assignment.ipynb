### Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. Write a Python function that creates a new column 'Username' in df that contains only the username part of each email address.

The username is the part of the email address that appears before the '@' symbol. For example, if the
email address is 'john.doe@example.com', the 'Username' column should contain 'john.doe'. Your
function should extract the username from each email address and store it in the new 'Username'
column.




```python
def create_column(df, new_column, old_column):
    element_list = []
    for i in df["old_column"]:
        xx = i.split("@")
        element_list.append(xx[0])
    df["new_column"] = element_list
    print(df["new_column"])
             
    
create_column(df, "Username", "Email")  
```

### Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.


```python
df.set_index('Date')

# Calculating the 7-day simple moving average of the values in 'Sales' column and assigning the corresponsing data to the 
# new column 'MovingAverage'

# Moving Average is calculating the average of data over a period of time. The moving average is also known as 
# the 'rolling mean'. The brolling() function creates a window of a specified size over the DataFrame and applies a 
# specified function to the values in the window. The size of the window determines the number of data points that are 
# used to calculate the moving average.

df['MovingAverage'] = df['Sales'].rolling(7).mean()

print(df['Moving Average'])
```

### Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g. Monday, Tuesday) corresponding to each date in the 'Date' column.

For example, if df contains the following values:


<pre>

  Date

0 2023-01-01

1 2023-01-02

2 2023-01-03

3 2023-01-04

4 2023-01-05

</pre>

Your function should create the following DataFrame:

<pre>

    Date     Weekday

0 2023-01-01 Sunday

1 2023-01-02 Monday

2 2023-01-03 Tuesday

3 2023-01-04 Wednesday

4 2023-01-05 Thursday

</pre>

The function should return the modified DataFrame.

```python
df7 = pd.DataFrame({
    'Date' : ['2023-01-01', '2023-01-02', '2023-01-03' ,'2023-01-04', '2023-01-05']
})

# 'to_datetime()' - This function converts a scalar, array-like, Series or DataFrame/dict-like to a pandas datetime object.
# in the below, converting all the element of the 'Date' column into datetime object.

df7['Date'] = pd.to_datetime(df7['Date'])

df7['Weekday'] = df7['Date'].dt.day_name()

df7

# Note: Above concept taken from: https://stackoverflow.com/questions/30222533/create-a-day-of-week-column-in-a-pandas-dataframe-using-python
```
