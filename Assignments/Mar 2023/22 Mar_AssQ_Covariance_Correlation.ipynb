## Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.


```python
import pandas as pd
import numpy as np

np.random.seed(10)

df_exam_results = pd.DataFrame({'Study Hours': np.random.randint(2,11,size=20),
                                'Score' : np.random.randint(60,95,size=20)
                               })

df_exam_results.head()
```

```python
# Pearson Correlation between the 'Study Hour' & 'Score' Feature in the dataset

df_exam_results.corr(method='pearson')
```

* Pearson Correlation between the 'Study Hour' & 'Score' Feature in the dataset is 0.368111. So, there is a position relation between the two features. And, the correlation between the two features are 'Moderate'.

## Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables. Suppose you have collected data on the amount of sleep individuals get each night and their overall job satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two variables and interpret the result.


```python
import numpy as np
import pandas as pd

np.random.seed(1000)

df_SJ = pd.DataFrame({'Sleep Hours': np.random.randint(4,8,size=10),
                      'Job Satisfaction' : np.random.randint(1,10,size=10)
                     })
df_SJ
```

```python
# Spearman Correlation between the 'Sleep Hour' and 'Job Satisfaction'

df_SJ.corr(method='spearman')
```

* Spearman Correlation between the 'Sleep Hour' and 'Job Satisfaction' is +0.14. So, those two varaible have positive relationship between them. If one increases, other increases. If, one decreases, other decrease.

## Q3. Suppose you are conducting a study to examine the relationship between the number of hours of exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation between these two variables and compare the results.


```python
import pandas as pd
import numpy as np

np.random.seed(1000)

df_health = pd.DataFrame({'Excercise/Week': np.random.randint(2,14,size=20),
                         'BMI' : np.random.randint(18,28,size=20)
                         })
df_health
```

```python
# Pearson Correlation between 'Excecise/Week' and 'BMI'

df_health.corr(method='pearson')
```

```python
# Spearman Correlation between 'Excecise/Week' and 'BMI'

df_health.corr(method='spearman')
```

* Both pearson and spearman correlation coefficient shows positive relationship between two features and they are pretty near to each other in value.

## Q5. A survey was conducted to examine the relationship between age and preference for a particular brand of soft drink. The survey results are shown below:



```python
import pandas as pd


df_survey = pd.DataFrame({'Age(Years)': [25,42,37,19,31,28],
                          'Soft Drink Preference' : ['Coke', 'Pepsi', 'Mountain Dew', 'Coke', 'Pepsi', 'Coke']
                         })

df_survey
```

```python
df_survey_encoded = df_survey.copy()


# Calculating the mean age per soft Drink type under 'Soft Drink Preference' feature in the dataset

mean_age_series = df_survey_encoded.groupby('Soft Drink Preference')['Age(Years)'].mean()

# replacing the categorical data points of 'Soft Drink Preference' feature with the mean calculated age per soft drink type

df_survey_encoded['Soft Drink Preference'] = df_survey_encoded['Soft Drink Preference'].map(mean_age_series)

# New Dataset will have both features as numerical data types.

df_survey_encoded.head()
```

```python
# Pearson Correlation between 'Age' and 'Soft Drink Preference' variable

df_survey_encoded.corr(method='pearson')
```
