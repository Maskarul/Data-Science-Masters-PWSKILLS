### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

<pre>

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

</pre>

Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y,c="#0cdf5f",marker="^")
plt.xlabel("X Axis Values")
plt.ylabel("Y Axis Values")
plt.title("Scatter Plot Example")
plt.show()
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Note: The below concept is taken from: https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html

```python
import numpy as np
import matplotlib.pyplot as plt

line1_x = np.array([0, 1, 2, 3, 4, 5])
line1_y = np.array([0, 100, 200, 300, 400, 500])

line2_x = np.array([0, 1, 2, 3, 4, 5])
line2_y = np.array([50, 20, 40, 20, 60, 70])

line3_x = np.array([0, 1, 2, 3, 4, 5])
line3_y = np.array([10, 20, 30, 40, 50, 60])

line4_x = np.array([0, 1, 2, 3, 4, 5])
line4_y = np.array([200, 350, 250, 550, 450, 150])

fig, axs = plt.subplots(2,2)

axs[0,0].plot(line1_x, line1_y)
axs[0,0].set_title('Axis [0, 0]')
axs[0,1].plot(line2_x, line2_y, 'tab:orange')
axs[0,1].set_title('Axis [0, 1]')
axs[1,0].plot(line3_x, line3_y, 'tab:green')
axs[1,0].set_title('Axis [1, 0]')
axs[1,1].plot(line4_x, line4_y, 'tab:red')
axs[1,1].set_title('Axis [1, 1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

<pre>

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

</pre>

Bar Plot:
>A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories. One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories.

Why it is used?
>A bar chart is used when you want to show a distribution of data points or perform a comparison of metric values across different subgroups of your data. From a bar chart, we can see which groups are highest or most common, and how other groups compare against the others.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

fig, ax = plt.subplots()

ax.bar(company,profit)
ax.set_ylabel("profit Counts")
ax.set_title("Bar Plot: Comapany VS Profit")

plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)


<b><u>Box Plot:</u></b>

A Box Plot is the visual representation of the statistical five number summary of a given data set.

A Five Number Summary includes:

1. Minimum
2. First Quartile
3. Median (Second Quartile)
4. Third Quartile
5. Maximum



<u><b>Why it is used?</b><u>

Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups. They are built to provide high-level information at a glance, offering general information about a group of data’s symmetry, skew, variance, and outliers.

```python
import numpy as np 
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

boxes = [box1, box2]

fig, ax = plt.subplots()

# boxplot function taken data in a single argument, if you have multiple data to plot in the box plot you will have to create
# a list or a tuple of those data and then pass it as an argument in the boxplot method. Here, we have created 'boxes' list for
# the same logic.

ax.boxplot(boxes, labels=["BOX1", "BOX2"])
plt.show()
```
