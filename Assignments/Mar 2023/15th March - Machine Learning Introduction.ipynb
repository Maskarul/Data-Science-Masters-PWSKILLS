### Q1: Explain the following with an example:


* <b>Artificial Intelligence (AI)</b>: 

> Definition: Artificial intelligence or AI recreates human intelligence and behaviour using algorithms, data, and models to develop machines to mimic human intelligence and behaviour.

> Example: Roomba, the smart robotic vacuum, uses AI to analyze the size of the room, obstacles, and pathways. Just like a human being taking this information into account, Roomba then retains this information and creates the most efficient route for vacuuming. 

* <b>Machine Learning (ML)</b>: 

>Definition: Algorithms that learn from structured data to predict outputs and discover patterns in that data.

>Example: Recommendations on streaming platforms like Netflix, Amazon Prime are another form of machine learning


* <b>Deep Learning (DL)</b>: 

>Definition: Algorithms based on highly complex neural networks that mimic the way a human brain works to detect patterns in large unstructured data sets.

>Example: Deep learning is used in virtual assistants such as Alexa and Siri, which use Natural Language Processing (NLP).

### Q2: What is supervised learning? List some examples of supervised learning.

* Definition: 

> Supervised machine learning learns patterns and relationships between input and output data. It is defined by its use of labeled data. A labeled data is a dataset that contains a lot of examples of Features and Target. Supervised learning uses algorithms that learn the relationship of Features and Target from the dataset. This process is referred to as Training or Fitting. 


* Example:

  1. Email Filtering: Supervised learning is commonly used in email filtering to classify incoming emails as spam or legitimate.
  2. Credit Scoring: In credit scoring, supervised learning is used to predict the creditworthiness of loan applicants. 

### Q3: What is unsupervised learning? List some examples of unsupervised learning.

* Definition:

> Unsupervised learning is the training of a machine using information that is neither classified nor labeled and allowing the algorithm to act on that information without guidance. Here the task of the machine is to group(cluster) unsorted information according to similarities, patterns, and differences without any prior training of data.

* Example: 

  1. Customer and market segmentation: Clustering algorithms can help group people that have similar traits(who spend more, who spend less) and create customer personas for more efficient marketing and targeting campaigns.

  2. Clinical cancer studies: Machine learning and clustering methods are used to study cancer gene expression data (tissues) and predict cancer at early stages.

### Q6: What is train, validation and test split? Explain the importance of each term.


* <b>Training Set</b>:

> The training set is the portion of the dataset reserved to fit the model. In other words, the model sees and learns from the data in the training set to directly improve its parameters. 

> To maximize model performance, the training set must be (i) large enough to yield meaningful results (but not too large that the model overfits) and (ii) representative of the dataset as a whole. This will allow the trained model to predict any unseen data that may appear in the future.

* <b>Validation Set</b>:

> The validation set is the set of data used to evaluate and fine-tune a machine learning model during training, helping to assess the models performance and make adjustments. 

> By evaluating a trained model on the validation set, we gain insights into its ability to generalize to unseen data. This assessment helps identify potential issues such as overfitting, which can have a significant impact on the models performance in real-world scenarios.


* <b>Test Set</b>:

> The test set is the set of data used to evaluate the final performance of a trained model. 

> It serves as an unbiased measure of how well the model generalizes to unseen data, assessing its generalization capabilities in real-world scenarios. By keeping the test set separate throughout the development process, we obtain a reliable benchmark of the models performance.

### Q7: How can unsupervised learning be used in anomaly detection?


* <b>What is an anomaly?</b>

> An anomaly is an instance that deviates significantly from the normal or expected behavior of the data. For example, an anomaly in a credit card transaction might indicate fraud, while an anomaly in a customer review might indicate a valuable insight. Therefore, it is important to define what constitutes an anomaly and how to measure its severity.


* <b>How to measure anomaly scores in Unsupervised Learning?</b>

> One of the common ways to measure anomaly scores is to use the distance or similarity between an instance and the rest of the data. For example, if you use k-means clustering to group the data into clusters, you can compute the distance between each instance and its assigned cluster center. The larger the distance, the more anomalous the instance. Alternatively, you can use density-based methods, such as local outlier factor (LOF), to compute the relative density of each instance compared to its neighbors. The lower the density, the more anomalous the instance.
