### Q4: Explain Bayes's theorem with an example.

<u>DEFINITION</u>:
Bayes theorem, in simple words, determines the conditional probability of event A given that event B has already occurred.

P(A|B)=P(A)P(B|A)/P(B)

In relevance of Data Science, Bayes's Theorem can be explained as "the mathematical rule that describes how to update a belief, given some evidence. In other words – it describes the act of learning."



* <b>The Equation</b>: Posterior = Prior x (Likelihood over Marginal probability)

* There are four parts in the above equation:

    * Posterior probability (updated probability after the evidence is considered)
    * Prior probability (the probability before the evidence is considered)
    * Likelihood (probability of the evidence, given the belief is true)
    * Marginal probability (probability of the evidence, under any circumstance)

<br>

* Why to use baye's Theorem?
> In simple word, there are many times when it is easy to compute P(B|A) than P(A|B) and sometimes it is easy to compute P(A|B) than P(B|A). We can easily use P(B|A) to calculate P(A|B) using Baye's Theorem and vice versa. 

<br>

<b><u>Example1</u></b>:
What is the probability of a couple have two girls given that they atleast have one girl?

Ans:

* B --> BOY, G --> GIRL

* sample set = {BB, GG, BG, GB}

* Here, have to find P(2G|atleast 1G) = (P(2G) * P(atleast 1G|2G))/P(atleast 1G)


* P(atleast 1G|2G) --> probability of the couple having atleast one girl when given that couple have two girls  --> this is easy
                 = 1, when u have 2 girls, you always have atleast one girl.                
* P(2G) = 1/4 (sample set can have 4 combination, where having GG(2 girls) occurs once)
* P(atleast 1G) = 3/4 (sample set can have 4 combination, where having atleast 1 girl (BG,GB,GG) occurs thrice)

* So now, P(2G|atleast 1G) = (1/4 * 1)/3/4 = (1/4)/(3/4) = 1/3



By the above example you can understand that ho using Baye's theorem, we easily calculated P(2G|atleast 1G).

### Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

#### <u>Confidence Interval</u>:
The confidence interval is the range of values that you expect your estimate to fall between a certain percentage of the time if you run your experiment again or re-sample the population in the same way.

The confidence level is the percentage of times you expect to reproduce an estimate between the upper and lower bounds of the confidence interval, and is set by the alpha value.

#### <u>Calculating a confidence interval: what you need to know</u>:

* If you want to calculate a confidence interval on your own, you need to know:

    1. The point estimate you are constructing the confidence interval for
    2. The critical values for the test statistic
    3. The standard deviation of the sample
    4. The sample size

* Example: 

> You survey 100 Brits and 100 Americans about their television-watching habits, and find that both groups watch an average of 35 hours of television per week. 95% confidence interval. The variance in the GB estimate is 100, while the variance in the USA estimate is 25. 

1. Point estimate: 

In the TV-watching example, the point estimate is the mean number of hours watched: 35.
    
2. Critical value: 

In the TV-watching survey, there are more than 30 observations and the data follow an approximately normal distribution (bell curve), so we can use the z distribution for our test statistics.
   
For a two-tailed 95% confidence interval, the alpha value is 0.025, and the corresponding critical value is 1.96.

This means that to calculate the upper and lower bounds of the confidence interval, we can take the mean ±1.96 standard deviations from the mean.
 
 
3. Standard deviation:

* sqrt(100) = 10 for the GB estimate.
* sqrt(25) = 5 for the USA estimate.


4. Sample size: 

In our survey of Americans and Brits, the sample size is 100 for each group.

### Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.

<pre>
Given,
n = 50 (n>30, so t test is not possible. We will have to take z test)

so assuming, n = 25 just to conduct t test. 

x = 6
s = 2.5
CL = 95%

</pre>

* H<sub>0</sub>: μ = 0 (drug is not effective, no weight change, so weight loss average is zero)
* H<sub>a</sub>: μ > 0 (drug is effective, weight reduced, so weight loss average is positive)

* This is one tail t test, to be specific left tail t test. As, CL = 0.95, α = 0.05, so critical t value with df=25-1=24 is 1.711.

* t statistics = (x - μ)/(s/sqrt(n)) = (6-0)/(2.5/sqrt(25)) = 6/(2.5/5) = 6/0.5 = 12

* Critical t value >> t statistics -- Thus we reject the Null Hypothesis.

* Conclusion: the drug is significantly effective at a 95% confidence level in weight reduction.

### Q12. A researcher is testing the effectiveness of two different teaching methods on student performance. Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82 with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance using a t-test with a significance level of 0.01.




```python
'''

from above example:

<pre>

s1 = 6    s2 = 5
n1 = 8    n2 = 12
x1 = 85   x2 = 82

α = 0.01(significance level)  ---> confidence level = 1 - α = 0.99 --- 99%

α/2 = 0.005

* H<sub>0</sub>: μ1 = μ2  --> μ1 - μ2 = 0 (Null Hypothesis --> No difference of performance between two group of students)
* H<sub>a</sub>: μ1 ≠ μ2  --> μ1 - μ2 ≠ 0 (Alternate Hypothesis --> There is difference of performance between two group of students) 
                          --> Two tails test

</pre>

'''

import scipy.stats as st
import math

# s1, s2 --> standard deviations
# n1, n2 --> sample sizes
# x1, x2 --> sample means

s1 = 6   
s2 = 5
n1 = 8   
n2 = 12
x1 = 85   
x2 = 82
α = 0.005

d_f = n1+n2-2

def t_test_hypothesis(s1,s2,n1,n2,x1,x2,α,d_f):
    
    # Read "https://stackoverflow.com/questions/60699836/how-to-use-norm-ppf" to understand the below.

    # considering two-tail distribution as estimating population mean is a range between min and max, we will use α/2 in func
    critical_t_value = st.t.ppf(1-α/2, df=d_f)
    print("Critical t Value: ", critical_t_value)
    
    # when population standard deviattion are unknown, we take point estimate of sample standard deviation as population
    # standard deviation in the formula to calculate standard error(SE)
    
    mean_diff = x1 - x2
    mu_diff = 0 # as per null hypothesis, there is no difference in population mean 
    se1 = s1/math.sqrt(n1) # standard error for sample 1
    print("SE1: ", se1)
    se2 = s2/math.sqrt(n2) # standard error for sample 2
    print("SE2: ", se2)
    sed = math.sqrt(se1*2+se2*2) # standard error difference between two samples
    print("SED: ", sed)
    
    t_statistics = (mean_diff - mu_diff)/sed
    print("t Statistics: ", t_statistics)
    
    if t_statistics<0:
        if critical_t_value>t_statistics:
            print("We accept the Null Hypothesis that they is no difference of performance between two group of students")
        else:
            print("We reject the Null Hypothesis that they is no difference of performance between two group of students")
    else:
        if critical_t_value<t_statistics:
            print("We accept the Null Hypothesis that they is no difference of performance between two group of students")
        else:
            print("We reject the Null Hypothesis that they is no difference of performance between two group of students") 
    

t_test_hypothesis(s1,s2,n1,n2,x1,x2,α,d_f)
```

### Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean of 65. Calculate the 90% confidence interval for the true population mean.


```python
# Sample size is 50, which is > 30 ----> we can use the z-test
# Given Confidence Interval or level -> 90%
# we are going to take point estimate of sampe standard deviation as population standard deviation while calculating 
# the standard error(SE) as sample standard deviation is not known. 

import scipy.stats as st
import math

# below four values are given in the question

sample_size = 50
sample_mean = 65
sample_standard_deviation = 8
CL = 0.95  #CL --> Confidence Level

def calculate_population_mean(sample_mean, sample_standard_deviation, sample_size, CL):
    
    # Read "https://stackoverflow.com/questions/60699836/how-to-use-norm-ppf" to understand the below.

    alpha = 1 - CL
    half_alpha = alpha/2  # considering two-tail distribution
    critical_z_score = st.norm.ppf(1-half_alpha)
    print("Critical z value: ", critical_z_score)
    
    # when population standard deviattion are unknown, we take point estimate of sample standard deviation as population
    # standard deviation in the formula to calculate standard error(SE)
    
    standard_error = sample_standard_deviation/math.sqrt(sample_size)
    print("Standard Error:", standard_error)
    population_mean_min = sample_mean - (critical_z_score*standard_error)
    population_mean_max = sample_mean + (critical_z_score*standard_error)
    print("Population mean will fall under {} to {} with confidence level of {}%".format(population_mean_min, population_mean_max, CL*100))


calculate_population_mean(sample_mean, sample_standard_deviation, sample_size, CL)


# to understand calculation concept, visit: http://www.southampton.ac.uk/~cpd/mean2.html
```
