### Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?

1. "loc" is label-based, which means that you have to specify rows and columns based on their row and column labels. "iloc" is integer position-based, so you have to specify rows and columns by their integer position values (0-based integer position).

2. loc[A:B] - A and B included. iloc[n:m] - n included but m excluded.

### Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and calculates the difference between them in days, hours, and minutes using Pandas time delta. The program should prompt the user to enter the dates and display the result.

```python
import pandas as pd

date1 = pd.Timestamp(input("Enter Date1 in YYYY-MM-DD format:"))
date2 = pd.Timestamp(input("Enter Date2 in YYYY-MM-DD format:"))

time_diff = pd.Timedelta(date2-date1)

print("Time Difference in days: ", time_diff.days)

print("Time Difference in hours: ", time_diff.total_seconds()/3600)

print("Time Difference in minutes: ", time_diff.total_seconds()/60)
```

### Q9. Write a python program that reads a csv file containing categorical data and converts a specified column into a categorical data type. The program should prompt the user to enter the file path, column name and category order and then display the sorted data.

Below code knowledge was taken from:

1. https://www.datacamp.com/tutorial/categorical-data
2. https://medium.com/swlh/categorical-data-in-pandas-9eaaff71e6f3
3. https://soulsinporto.medium.com/custom-sort-a-pandas-dataframe-with-pd-categorical-c4eec8343957

```python
import pandas as pd

# using this csv data in this program: https://raw.githubusercontent.com/pycaret/pycaret/master/datasets/diamond.csv

path = input("Please enter the csv file path: ")

df_from_csv = pd.read_csv(path)

df_from_csv.head()

column_name = input("Please enter the column name within these {} which you wanted to be converted as a category: ".format(df_from_csv.columns))

# As I took a complex data, so for execution purpose of the pprogram, we are always going to choose "Color" as the column
# which we wanted to convert type of it as category.

df_from_csv[column_name] = df_from_csv[column_name].astype("category")

df_from_csv[column_name].unique()

sort_order_number = input('''
                             1 --> ['D', 'E', 'F', 'G', 'H', 'I']
                             2 --> ['E', 'D', 'G', 'F', 'I', 'H']
                             3 --> ['I', 'H', 'G', 'F', 'E', 'D']  
                             
                             Enter the number from the above to choose your desired sorting order:
                             
                             ''')

if sort_order_number == '1':
    sort_order = ['D', 'E', 'F', 'G', 'H', 'I']
elif sort_order_number == '2':
    sort_order = ['E', 'D', 'G', 'F', 'I', 'H']
else:
    sort_order = ['I', 'H', 'G', 'F', 'E', 'D']
    
# choosing the sort order for the column

df_from_csv[column_name] = pd.Categorical(df_from_csv[column_name], categories= sort_order)

# now sorting the dataframe as per the sorting order of the categorical column

df_from_csv.sort_values(by = column_name)
```

### Q10. Write a python program that reads a csv file containing sales data from different products and visualizes the data using a stacked bar chart to show the sales of each product category over time. The program should prompt the user to entter the file ath and display the chart.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sales_df = pd.read_csv('Sales_Data.csv')

sales_df.drop('Unnamed: 0', axis = 1, inplace=True)

display(sales_df)

months = list(sales_df['months'])
jeans = list(sales_df['jeans'])
tshirt = list(sales_df['tshirt'])
formal_shirt = list(sales_df['formal_shirt'])

#add colors
colors = ['#FF9999', '#00BFFF','#C1FFC1','#CAE1FF','#FFDEAD']

# The position of the bars on the x-axis
r = range(len(months))

barWidth = 1
#plot bars
plt.figure(figsize=(8,5))
plt.bar(r, jeans, color=colors[0], edgecolor='white', width=barWidth, label="jeans")
plt.bar(r, tshirt, bottom=np.array(jeans), color=colors[1], edgecolor='white', width=barWidth, label='tshirt')
plt.bar(r, formal_shirt, bottom=np.array(jeans)+np.array(tshirt), color=colors[2], edgecolor='white', width=barWidth, label='formal shirt')
plt.legend()# Custom X axis
plt.xticks(r, months, fontweight='bold')
plt.ylabel("sales")
plt.savefig("stacked1.png")
plt.show()
```

### Q11. You are given a CSV file containing student data that includes Student ID and their test scores. Write a python programs that reads the CSV file, calculates mean, median, and the mode of the test scores, and displays the results in a table.

The program should be doing the following:

1. Prompt the user to enter the path of the csv file containing the student data.
2. Read the csv file into a pandas dataframe
3. Calculate the mean, median and mode of the test scores using pandas tools.
4. Display the mean, median and mode in a table.



```python
import pandas as pd

path = input("Enter the path of the file: ")

student_test_score_df = pd.read_csv(path)

display(student_test_score_df)

mean = student_test_score_df['Test Score'].mean()
median = student_test_score_df['Test Score'].median()
mode = student_test_score_df['Test Score'].mode()


print('''+-----------+---------+
           | Statistic |   Value |
           +-----------+---------+
           | Mean      |   {}    |
           | Median    |   {}    |
           | Mode      |   {}    |
           |           |         |
           +-----------+---------+
           '''.format(mean, median, mode))
```
